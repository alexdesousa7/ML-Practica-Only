# Exploración de Datos
Después de limpieza y transformación de datos, vamos explorar el estado actual de los mismos.

## Importación de Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

pd.set_option("display.max_columns", 100)
sns.set(style="whitegrid")
```

## Definicion de variables

```python
PATH_DIRECTORIO_DATOS = "./../../data/processed"
PATH_DATASET = f"{PATH_DIRECTORIO_DATOS}/processed_dataset.csv"
```

## Importar dataset

```python
df = pd.read_csv(PATH_DATASET)
df.head()
```

# Analisis de la variable objetivo
A pesar de las transformaciones, los datos deben permancer igual en terminos de porcentajes y distribución.

```python
total = len(df)
canceled = df["is_canceled"].sum()
not_canceled = total - canceled

canceled_pct = (canceled / total) * 100
not_canceled_pct = (not_canceled / total) * 100

print(f"Canceladas:     {canceled} ({canceled_pct:.1f}%)")
print(f"No canceladas:  {not_canceled} ({not_canceled_pct:.1f}%)")
print(f"Total:          {total}")
```

# Revisando que los datos son correctos
- Variables eliminadas correctamente.
- Nuevas variables agregadas.
- Validando que las variables positivas NO tengan valores negativos.

```python
dropped_cols = ["reservation_status", "reservation_status_date", "company", "agent", "lead_time", "reserved_room_type", "assigned_room_type"]
new_cols     = ["room_mismatch", "lead_time_log", "has_agent"]
positive_cols = ["adr", "days_in_waiting_list", "stays_in_week_nights", "stays_in_weekend_nights", "total_of_special_requests"]

print("=" * 50)
print("1. Columnas eliminadas")
print("=" * 50)
for col in dropped_cols:
    status = "TODAVÍA EXISTE" if col in df.columns else "Eliminada"
    print(f"  {status}: {col}")

print("\n" + "=" * 50)
print("2. Columnas nuevas (feature engineering)")
print("=" * 50)
for col in new_cols:
    status = "✓ Existe" if col in df.columns else "NO EXISTE"
    print(f"  {status}: {col}")

print("\n" + "=" * 50)
print("3. Valores negativos")
print("=" * 50)
for col in positive_cols:
    negatives = (df[col] < 0).sum()
    status = "Sin negativos" if negatives == 0 else f"✗ {negatives} NEGATIVOS ENCONTRADOS"
    print(f"  {status}: {col}")

print("\n" + "=" * 50)
print("4. Resumen del dataset procesado")
print("=" * 50)
print(f"  Filas:    {df.shape[0]:,}")
print(f"  Columnas: {df.shape[1]}")
print(f"  Nulos:    {df.isnull().sum().sum()}")
```
