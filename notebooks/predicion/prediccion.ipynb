```python
# 1._ Importar librerias y cargar modelo
```

```python
import pandas as pd
import joblib

# Cargar modelo final
model = joblib.load("../../models/best_model.pkl")

```

```python
# 2._ Cargar datos de ejemplo para predecir
```

```python
df = pd.read_csv("../../data/raw/dataset_practica_final.csv")

# Tomamos una fila cualquiera
sample = df.sample(1, random_state=42)

sample
```

```python
# 3._ Separar X y aplicar prediccion
```

```python
X_sample = sample.drop("is_canceled", axis=1)

pred = model.predict(X_sample)[0]
prob = model.predict_proba(X_sample)[0][1]

print("Predicción:", pred)
print("Probabilidad de cancelación:", prob)

```

```python
# 4._ Interpretar resultados
```

```python
if pred == 1:
    print("El modelo predice que la reserva será CANCELADA.")
else:
    print("El modelo predice que la reserva NO será cancelada.")

```
